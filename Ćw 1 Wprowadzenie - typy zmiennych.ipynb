## Zmienne

Zmienna — wydzielone miejsce w pamięci komputera, o określonej nazwie, w którym przechowywane są dane.
Przypisanie wartości do zmiennej:
- nazwa = ’zawartość’
- =   operator przypisania
- ważne funkcje: print(), type()

```python
zmienna = 5
print(zmienna)
```

Nazwy zmiennych - podstawowe zasady:
- nie mogą zaczynać się od cyfry, np. 3krol jest nieprawidłową nazwą
- nie mogą zawierać spacji, zamiast nich używamy podkreślnika '_'
- zwyczajowo zaczynamy od małej litery
- unikamy używania zwrotów, które mają przypisane inne znaczenie np. print, list, str. Środowisko programistyczne powinno zaznaczyć takie słowo kolorem
- nie mogą zawierać następujących symboli: :'",<>/?\()!@#$%^&*~-+

Konwencja nazywania zmiennych:
- snake_case — małe litery i podkreslniki, aby rozdzielić wyrazy np my_name. Preferowany dla funkcji, metod, atrybutów i zmiennych

Dodatkowo:
ALLCAPS - wszystko dużymi literami, stosowane dla stałych.
camelCase — zaczynamy małą literą, następne wyrazy z dużej np myName. Tylko, jeżeli już jest używany w kodzie.
StudlyCaps — dla klas np MyName


Python ma dynamiczne typowanie zmiennych — oznacza to bieżące przypisywanie typów do wartości przechowywanych w zmiennych w trakcie działania programu. Dzięki temu programowanie jest prostsze i szybsze, jednak może to być przyczyną błędów. Warto korzystać z funkcji type (), która podaje typ zmiennej

```python
# Typowanie dynamiczne
a = 5
print(type(a))

a = "rycerz"
print(a)
print(type(a))
```

```python
a = 2
# noinspection PyStatementEffect
a + a
# Operator + zawsze ma pierwszeństwo przed =, dlatego poniższy kod działa
a = a + a + a
print(a)
```

## Tekst
Typ tekstowy (ang. String), należy do sekwencji. Skrót str


```python
tekst1 = 'Jaka jest prędkość lotu jaskółki bez obciążenia?'
tekst2 = "Jakiej 'jaskółki' afrykańskiej czy europejskiej?"
print(tekst1, tekst2)
```

Typ tekstowy jest sekwencją i jest uporządkowany, oznacza to że może być indeksowany z użyciem [] i dzielony na fragmenty

```python
# Indeksowanie - wybieranie fragmentu z ciągu znaków. Ważne! Numerowanie pozycji zaczyna się od 0!
tekst1[1]
```

```python
# Indeksowanie w tył - tutaj pierwsze od tyłu jest -1
tekst1[-1]
```

#### Ćwiczenie 3
Opisz w komentarzu co dzieje się w poniższych sytuacjach, jakie fragmenty tekstu są wybierane?

```python
# Wycinanie(slicing) wg wzoru [start:stop:step]
print(tekst1[0:4])        # pierwsze 4 znaki
print(tekst1[:4])
print(tekst1[5:])
print(tekst1[::3])
print(tekst1[1:8:2])
print(tekst1[::-1])
```

```python
# Długość sekwencji
len(tekst1)
```

```python
# Łączenie tekstu
print('2'+'3')

# Znaki tabulacji i nowej linii
print('Strażnik:\t' + tekst1 + '\nKról Artur:\t' + tekst2)
```

```python
# print(2 + '3')
# Ważne! Nie można łączyć tekstu z liczbą, trzeba najpierw zmienić jej typ na tekst za pomocą funkcji str()
# Jest to konwersja wartości zmiennej (ang. Type Casting) czyli zmiana typu.
# Wyróżniamy konwersję automatyczną dokonywaną "w locie" (ang. implicit) oraz wykonywaną za pomocą specjalnych funkcji (ang. explicit)
print(str(2)+'3')
```

## Najważniejsze metody klasy string

```python
# Po naciśnięciu tabulatora wyświetlą się dostępne metody
# print(tekst1.)

# Przykłady
print(dna.count('A'))
print(dna.lower())
print(tekst1.title())
```
