## Zmienne

Zmienna  wydzielone miejsce w pamici komputera, o okrelonej nazwie, w ktrym przechowywane s dane.
Przypisanie wartoci do zmiennej:
- nazwa = zawarto
- =   operator przypisania
- wane funkcje: print(), type()

```python
zmienna = 5
print(zmienna)
```

Nazwy zmiennych - podstawowe zasady:
- nie mog zaczyna si od cyfry, np. 3krol jest nieprawidow nazw
- nie mog zawiera spacji, zamiast nich uywamy podkrelnika '_'
- zwyczajowo zaczynamy od maej litery
- unikamy uywania zwrotw, ktre maj przypisane inne znaczenie np. print, list, str. rodowisko programistyczne powinno zaznaczy takie sowo kolorem
- nie mog zawiera nastpujcych symboli: :'",<>/?\()!@#$%^&*~-+

Konwencja nazywania zmiennych:
- snake_case  mae litery i podkreslniki, aby rozdzieli wyrazy np my_name. Preferowany dla funkcji, metod, atrybutw i zmiennych

Dodatkowo:
ALLCAPS - wszystko duymi literami, stosowane dla staych.
camelCase  zaczynamy ma liter, nastpne wyrazy z duej np myName. Tylko, jeeli ju jest uywany w kodzie.
StudlyCaps  dla klas np MyName


Python ma dynamiczne typowanie zmiennych  oznacza to biece przypisywanie typw do wartoci przechowywanych w zmiennych w trakcie dziaania programu. Dziki temu programowanie jest prostsze i szybsze, jednak moe to by przyczyn bdw. Warto korzysta z funkcji type (), ktra podaje typ zmiennej

```python
# Typowanie dynamiczne
a = 5
print(type(a))

a = "rycerz"
print(a)
print(type(a))
```

```python
a = 2
# noinspection PyStatementEffect
a + a
# Operator + zawsze ma pierwszestwo przed =, dlatego poniszy kod dziaa
a = a + a + a
print(a)
```

## Tekst
Typ tekstowy (ang. String), naley do sekwencji. Skrt str

Typ tekstowy jest sekwencj i jest uporzdkowany, oznacza to e moe by indeksowany z uyciem [] i dzielony na fragmenty

```python
# Indeksowanie - wybieranie fragmentu z cigu znakw. Wane! Numerowanie pozycji zaczyna si od 0!
tekst1[1]
```

```python
# Indeksowanie w ty - tutaj pierwsze od tyu jest -1
tekst1[-1]
```
